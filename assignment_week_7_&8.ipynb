```python
#week 7

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Math': [78, 55, 88, 62, 75, 92, 68, 85, 70, 95],
    'Science': [80, 60, 90, 65, 78, 94, 70, 88, 72, 96],
    'English': [65, 72, 76, 80, 70, 85, 78, 82, 68, 90],
    'Computer': [82, 58, 85, 68, 79, 90, 75, 88, 72, 93]
}

df_college = pd.DataFrame(data)

# Calculate the correlation matrix
corr_college = df_college.corr()

# Print the correlation matrix
print("Correlation Matrix:\n", corr_college)

# Visualize the correlation matrix with a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_college, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap of Subject Scores')
plt.show()
```

```python
#week 8

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'BMI': [22.5, 27.8, 31.0, 24.5, 29.2, 26.8, 33.5, 21.7, 25.4, 28.1],
    'Sleep_Hours': [7, 6, 5, 8, 6, 7, 4, 8, 7, 5],
    'Physical_Activity': [60, 30, 15, 90, 20, 45, 10, 120, 70, 25],
    'Sugar_Level': [95, 110, 140, 90, 130, 100, 160, 85, 92, 125],
    'Blood_Pressure': [115, 130, 145, 110, 140, 120, 150, 105, 118, 135]
}
df = pd.DataFrame(data)
corr = df.corr()
print("Correlation Matrix:\n", corr)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.show()
```
